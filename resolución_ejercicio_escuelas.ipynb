Para este ejercicio vamos a importar un conjunto de datos con el listado de colegios públicos de Bogotá del año 2016

# Ejercicios Pandas I

```python
import pandas as pd
```

```python
df_colegios = pd.read_csv('https://www.datos.gov.co/api/views/qijw-htwa/rows.csv?accessType=DOWNLOAD')
```

¿Cuáles son las columnas? ¿Cuántas filas tiene?

```python
df_colegios.columns
```

```python
df_colegios.shape
```

```python
df_colegios.sample(5)
```

Seleccionen los colegios donde la propiedad de la planta física sea de una persona natural o comunidad religiosa.

```python
df_colegios.internado.unique
```

```python
df_colegios.query("propiedad_Planta_Fisica == 'PERSONA NATURAL' or propiedad_Planta_Fisica == 'COMUNIDAD RELIGIOSA'")
```

¿Hay alguno de estos colegios que sea para capacidades excepcionales?

```python
df_colegios.pivot_table(
    index = ["capacidades_Excepcionales"],
    values = ["nombreestablecimiento"],
    aggfunc = "count"
)
```

```python
df_colegios["capacidades_Excepcionales"].unique()
```

¿Cuáles son los colegios que ofrecen todos los niveles educativos? ¿Qué idiomas ofrecen?

```python
df_colegios.query("capacidades_Excepcionales == 'CAPACIDADES EXCEPCIONALES'").count()
```

```python
df_colegios.pivot_table(
    index = ["niveles"],
    values = ["nombreestablecimiento"],
    aggfunc = "count"
)
```

```python
df_colegios.pivot_table(
    index = ["idiomas", "niveles"],
    values = ["nombreestablecimiento"],
    aggfunc = "count"
)
```
